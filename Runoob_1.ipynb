```python
quick_sort <- function(arr) {
  if (length(arr) <= 1) {
    return(arr)
  } else {
    pivot <- arr[length(arr) %/% 2 + 1]
    left <- arr[arr < pivot]
    middle <- arr[arr == pivot]
    right <- arr[arr > pivot]
    return(c(quick_sort(left), middle, quick_sort(right)))
  }
}

# Example usage:
arr <- c(3, 6, 8, 10, 1, 2, 1)
sorted_arr <- quick_sort(arr)
print("Sorted array:")
print(sorted_arr)
```

```python
import random
from IPython.display import clear_output  # 可选：用于清屏刷新界面（体验更佳）

def show_ascii_art(choice):
    """显示对应选择的 ASCII 图像"""
    arts = {
        "石头": """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
""",
        "剪刀": """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
""",
        "布": """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""
    }
    print(arts.get(choice, ""))
```

```python
def play_rock_paper_scissors():
    choices = ["石头", "剪刀", "布"]
    win_count = 0
    lose_count = 0
    tie_count = 0

    print("🎮 欢迎来到【石头剪刀布】游戏！")
    print("=" * 50)

    while True:
        # 显示选项图像
        print("\n请选择你的出拳（输入对应数字）：")
        print("1 → 石头")
        show_ascii_art("石头")
        print("2 → 剪刀")
        show_ascii_art("剪刀")
        print("3 → 布")
        show_ascii_art("布")
        print("0 → 退出游戏\n")

        user_input = input("请输入你的选择 (0/1/2/3)：").strip()

        if user_input == "0":
            print(f"\n🎉 游戏结束！总战绩 → 胜: {win_count} | 负: {lose_count} | 平: {tie_count}")
            print("👋 谢谢游玩！再见！")
            break

        if user_input not in ["1", "2", "3"]:
            print("❌ 无效输入！请重新输入 0、1、2 或 3。")
            continue

        # 映射数字到选择
        user_choice = choices[int(user_input) - 1]
        computer_choice = random.choice(choices)

        print("\n" + "="*30)
        print("🎮 你出的是：")
        show_ascii_art(user_choice)
        print(f"→ {user_choice}")

        print("\n🤖 计算机出的是：")
        show_ascii_art(computer_choice)
        print(f"→ {computer_choice}")
        print("="*30)

        # 判断胜负
        if user_choice == computer_choice:
            print("🤝 平局！")
            tie_count += 1
        elif (user_choice == "石头" and computer_choice == "剪刀") or \
             (user_choice == "剪刀" and computer_choice == "布") or \
             (user_choice == "布" and computer_choice == "石头"):
            print("🏆 你赢了！")
            win_count += 1
        else:
            print("😭 你输了！")
            lose_count += 1

        print(f"📊 当前战绩 → 胜: {win_count} | 负: {lose_count} | 平: {tie_count}")
        print("-" * 50)

# 启动游戏
play_rock_paper_scissors()
```

```python
def my_func(x, y, z=10):
    return x + y + z

def wrapper(func, *args, **kwargs):
    # do something here (if needed)
    return func(*args, **kwargs)

# 使用 wrapper 调用
result = wrapper(my_func, 1, 2, z=5)
print(result)  # 输出 8
```

```python
# 如何在外部调用嵌套函数（内层函数
# 可以通过让 outer() 返回 inner 函数对象，从而在外部拿到并调用它 —— 这就是 闭包（Closure） 的核心思想！ 
# 强烈推荐  Python 闭包标准用法，清晰安全
def outer():
    def inner():
        print("Hello from inner!")

    return inner  # ✅ 返回函数对象，不是调用！<<========================

# 在外部拿到 inner
f = outer()   # 调用 outer，拿到 inner 函数对象
f()           # 调用 inner → 输出: Hello from inner! <<=======
```

```python
# ：使用类封装（更面向对象）
# 推荐 更适合复杂场景或需多次调用
class Outer:
    def __init__(self):
        pass

    def inner(self):
        print("Inner method called-------------")

# 外部调用
obj = Outer()
obj.inner()  # ✅ 正常调用
```

```python
# 闭包进阶示例：带状态的内层函数
def outer(x):
    def inner(y):
        return x + y  # inner 记住了 outer 的 x
    return inner

add5 = outer(5**2)   # 返回一个“记住 x=5”的函数
print(add5(3))    # 输出: 8
print(add5(10))   # 输出: 15
```
