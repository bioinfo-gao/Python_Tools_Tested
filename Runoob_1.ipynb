```python
quick_sort <- function(arr) {
  if (length(arr) <= 1) {
    return(arr)
  } else {
    pivot <- arr[length(arr) %/% 2 + 1]
    left <- arr[arr < pivot]
    middle <- arr[arr == pivot]
    right <- arr[arr > pivot]
    return(c(quick_sort(left), middle, quick_sort(right)))
  }
}

# Example usage:
arr <- c(3, 6, 8, 10, 1, 2, 1)
sorted_arr <- quick_sort(arr)
print("Sorted array:")
print(sorted_arr)
```

```python
# å¦‚ä½•åœ¨å¤–éƒ¨è°ƒç”¨åµŒå¥—å‡½æ•°ï¼ˆå†…å±‚å‡½æ•°
# å¯ä»¥é€šè¿‡è®© outer() è¿”å›ž inner å‡½æ•°å¯¹è±¡ï¼Œä»Žè€Œåœ¨å¤–éƒ¨æ‹¿åˆ°å¹¶è°ƒç”¨å®ƒ â€”â€” è¿™å°±æ˜¯ é—­åŒ…ï¼ˆClosureï¼‰ çš„æ ¸å¿ƒæ€æƒ³ï¼ 
# å¼ºçƒˆæŽ¨è  Python é—­åŒ…æ ‡å‡†ç”¨æ³•ï¼Œæ¸…æ™°å®‰å…¨
def outer():
    def inner():
        print("Hello from inner!")

    return inner  # âœ… è¿”å›žå‡½æ•°å¯¹è±¡ï¼Œä¸æ˜¯è°ƒç”¨ï¼<<========================

# åœ¨å¤–éƒ¨æ‹¿åˆ° inner
f = outer()   # è°ƒç”¨ outerï¼Œæ‹¿åˆ° inner å‡½æ•°å¯¹è±¡
f()           # è°ƒç”¨ inner â†’ è¾“å‡º: Hello from inner! <<=======
```

```python
# é—­åŒ…è¿›é˜¶ç¤ºä¾‹ï¼šå¸¦çŠ¶æ€çš„å†…å±‚å‡½æ•°
def outer(x):
    def inner(y):
        return x + y  # inner è®°ä½äº† outer çš„ x
    return inner

add5 = outer(5**2)   # è¿”å›žä¸€ä¸ªâ€œè®°ä½ x=5â€çš„å‡½æ•°
print(add5(3))    # è¾“å‡º: 8
print(add5(10))   # è¾“å‡º: 15
```
