```python
class Duck:
    def quack(self):
        print('Quaaack!')

    def walk(self):
        print('Walks like a duck.')

def main():
    donald = Duck()
    print(donald)
    donald.quack()
    donald.walk()

if __name__ == "__main__": main()
```

```python
class Duck:
    def __init__(self):
        print('New Duck created')
        print('Duck Constructor')
    def quack(self):
        print('Quaaack!')

    def walk(self):
        print('Walks like a duck.')

def main():
    donald = Duck() # the constructor is called
    print(donald)
    donald.quack()
    donald.walk()

if __name__ == "__main__": main()
```
