```python
class Duck:
    def quack(self):
        print('Quaaack!')

    def walk(self):
        print('Walks like a duck.')

def main():
    donald = Duck()
    print(donald)
    donald.quack()
    donald.walk()

if __name__ == "__main__": 
    main()
```

```python
class Duck:
    def __init__(self):
        print('New Duck created')
        print('Duck Constructor')
    def quack(self):
        print('Quaaack!')

    def walk(self):
        print('Walks like a duck.')

def main():
    donald = Duck() # the constructor is called
    print(donald)
    donald.quack()
    donald.walk()

if __name__ == "__main__": 
    main()
```

```python
class Duck:
    def __init__(self, value):
        self.value = value
        print('New Duck created with ${self.value}')
        print(f'The fstring method ${self.value}')                  # The correct Python f-string syntax 
        print('Duck Constructor with ${}'.format(self.value))
    def quack(self):
        print('Quaaack!', self.value)

    def walk(self):
        print('Walks like a duck.', self.value)

def main():
    donald = Duck(47) # the constructor is called
    print(donald)
    donald.quack()
    donald.walk()

if __name__ == "__main__": 
    main()
```
